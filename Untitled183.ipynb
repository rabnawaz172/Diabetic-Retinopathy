```python
import shutil
import os
from google.colab import drive


# Define the output directory in Google Drive
output_dir = '/content/drive/MyDrive/Mango_Models'



```

```python
# prompt: check how many files found in directory

# Check if the output directory exists, if not create it
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Count the number of files in the directory
num_files = len([f for f in os.listdir(output_dir) if os.path.isfile(os.path.join(output_dir, f))])

print(f"Number of files in '{output_dir}': {num_files}")
```

```python
import os
import matplotlib.pyplot as plt
import json
from google.colab import files

# Function to plot metrics for multiple models
def plot_metric(ax, metric_key, histories, metric_name):
    for model, (history, model_name) in histories.items():
        ax.plot(
            range(1, len(history[metric_key]) + 1),
            history[metric_key],
            label=model_name,
            linewidth=2  # Solid lines with consistent width
        )
    ax.set_title(metric_name, fontsize=20)
    ax.set_xlabel("Epochs", fontsize=18)
    ax.set_ylabel(metric_name, fontsize=18)
    ax.legend(fontsize=14)
    ax.grid(alpha=0.3)
    ax.tick_params(axis="both", labelsize=16)

# Set the directory where history files are located


# List of model names in the specified order
models = [("cnn", "A)CNN"), ("efficientnet", "B)EfficientNet"), ("resnet", "C)ResNet"), ("densenet", "D)DenseNet")]
histories = {}

# Check and load each model's history
for model, name in models:
    history_path = os.path.join(output_dir, f"{model}_training_history.json")
    if not os.path.exists(history_path):
        print(f"File not found: {history_path}")
        continue
    with open(history_path, 'r') as f:
        histories[model] = (json.load(f), name)

# Ensure all histories are loaded
if len(histories) != len(models):
    print("Some history files are missing. Loaded histories:", list(histories.keys()))

# Create a 2x2 figure for plots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot training accuracy
plot_metric(axs[0, 0], "train_accuracy", histories, "A) Training Accuracy")

# Plot validation accuracy
plot_metric(axs[0, 1], "val_accuracy", histories, "B) Validation Accuracy")

# Plot training loss
plot_metric(axs[1, 0], "train_loss", histories, "C) Training Loss")

# Plot validation loss
plot_metric(axs[1, 1], "val_loss", histories, "D) Validation Loss")

# Adjust layout and save the figure with 400 DPI
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the title
plot_path = '/content/all_accuracy_loss_comparision_epochs.png'  # Save to /content directory
plt.savefig(plot_path, dpi=400)  # Set DPI to 400 for high-resolution output
plt.show()

# Download the plot file
files.download(plot_path)

print(f"Combined metrics plot saved and ready for download: {plot_path}")
```
